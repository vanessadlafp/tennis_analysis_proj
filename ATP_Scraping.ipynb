```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd

base_site = 'https://www.atptour.com/en/rankings/singles?rankDate=2019-12-30&rankRange=0-100'
r = requests.get(base_site , headers={'User-Agent': ''})
r
```

```python
html= r.content
```

```python
soup= BeautifulSoup(html, "lxml")
```

```python
df_players_2019= pd.read_html(r.text)[0]
df_players_2019
```

```python
player_urls_2019=[]

for i in range(df_players_2019.shape[0]):
    site= 'https://www.atptour.com' +soup.select('.player-cell a')[i]['href']
    player_urls_2019.append(site[:-8] +'player-stats?year=2019&surfaceType=all')                                                 
player_urls_2019                                                     
```

```python
stats_cols= ['Aces', 'Double Faults', '1st Serve','1st Serve Points Won', '2nd Serve Points Won', 'Break Points Faced',
       'Break Points Saved', 'Service Games Played', 'Service Games Won', 'Total Service Points Won', '1st Serve Return Points Won',
       '2nd Serve Return Points Won', 'Break Points Opportunities', 'Break Points Converted','Return Games Played', 'Return Games Won',
       'Return Points Won', 'Total Points Won']
```

```python
Turned_Pro =[]
stats_all= pd.DataFrame(columns=stats_cols)

for url in player_urls_2019: 
    player_page = requests.get(url, headers={'User-Agent': ''})
    player_html=player_page.content
    player_soup = BeautifulSoup(player_html, 'lxml')
    
    stats_tables = pd.read_html(player_page.text)
    Turned_Pro.append(stats_tables[0][1][0])
    
    numbers=pd.concat((stats_tables[1].iloc[:, 1],stats_tables[2].iloc[:,1]), axis=0).T
    df = pd.DataFrame([[num for num in numbers]], columns=stats_cols)
    stats_all=pd.concat([stats_all, df])
```

```python
stats_all=stats_all.reset_index()
stats_all
```

```python
turned_pro= [year.strip('Turned Pro  ') for year in Turned_Pro]
turned_pro
```

```python
df_players_2019['Turned Pro']= turned_pro
df_players_2019
```

```python
df_players_2019= pd.concat((df_players_2019,stats_all), axis=1)
df_players_2019
```
